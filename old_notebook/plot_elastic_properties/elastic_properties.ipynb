# Elastic Properties

```python
import tkinter as tk
from tkinter import Tk
from tkinter.filedialog import askopenfilename,askdirectory,asksaveasfilename
from CRYSTALpytools.crystal_io import Crystal_output
import CRYSTALpytools.plot as cfplt
```

Choose CRYSTAL file

```python
Tk().withdraw()
filename = askopenfilename()
data = Crystal_output().read_cry_output(filename)
tensor = data.get_elatensor()
```

Set a ndeg value

```python
ndeg = 200
```

## Young Modulus

```python
cfplt.plot_cry_ela("young", ndeg, tensor)
```

## Linear Compressibility

```python
cfplt.plot_cry_ela("comp", ndeg, tensor)
```

## Shear Modulus

### Average

```python
cfplt.plot_cry_ela("shear avg", ndeg, tensor)
```

### Minimum

```python
cfplt.plot_cry_ela("shear min", ndeg, tensor)
```

### Maximum

```python
cfplt.plot_cry_ela("shear max", ndeg, tensor)
```

```python
cfplt.plot_cry_ela("poisson avg", ndeg, tensor)
```
