# Transport Properties: A Quick Overview

CRYSTALpytools contains functions to read the computed electron transport properties from CRYSTAL properties output file and plot them. Please, note that these file can be generated thanks to the BLOTZTRA option (see the User's Manusal for details). Below, we will provide examples on how to exploit such functionalities.

As for all the CRYSTALpytools methods and functions, at first they need to be imported:

```python
from CRYSTALpytools.crystal_io import *
from CRYSTALpytools.plot import *
```

## Seebeck Coeffiecient

In CRYSTAL the Seebeck coefficient ($S$) of a material can be computed. It is a measure of the magnitude of an induced thermoelectric voltage in response to a temperature difference across the material, as induced by the Seebeck effect. It can be read from a CRYSTAL properties output file and transformed into a CRYSTAL object as follows:

```python
seebeck = Properties_output().read_cry_seebeck('../data/bi2te3_SEEBECK.DAT')
```

The Seebeck coefficient can be plotted with respect to the chemical potential or to the charge carrier concentration with the two functions repoted below. The functions will ask as an input the direction in which you want the Seebeck coefficent to be plotted. To differentiate transport coefficients coming from a p-type or an n-type conduction, solid and dashed lines are used respectively.

```python
plot_cry_seebeck_potential(seebeck)
```

```python
plot_cry_seebeck_carrier(seebeck)
```

You can also plot multiple Seebeck coefficients from multiple file with respect to the chemical potential, as shown below. The function will ask you as an input the temperature, the chemical potential range and the direction you want to plot.

```python
seebeck1 = Properties_output().read_cry_seebeck('../data/tinisn_SEEBECK.DAT')
```

```python
plot_cry_multiseebeck(seebeck, seebeck1)
```

## Electron Conductivity

In CRYSTAL the electron conductivity ($\sigma$) of a material can be computed. It represents a material's ability to conduct electric current. It can be read from a CRYSTAL properties output file and transformed into a CRYSTAL object as follows:

```python
sigma = Properties_output().read_cry_sigma('../data/bi2te3_SIGMA.DAT')
```

The electron conductivity can be plotted with respect to the chemical potential or to the charge carrier concentration with the two functions repoted below. The functions will ask as an input the direction in which you want the electron conductivity to be plotted.

```python
plot_cry_sigma_potential(sigma)
```

```python
plot_cry_sigma_carrier(sigma)
```

You can also plot multiple electron conductivities from multiple file with respect to the chemical potential, as shown below. The function will ask you as an input the temperature, the chemical potential range and the direction you want to plot.

```python
sigma1 = Properties_output().read_cry_sigma('../data/tinisn_SIGMA.DAT')
```

```python
plot_cry_multisigma(sigma, sigma1)
```

## Power Factor

The power factor of a material can be computed from its Seebeck coefficent and electron conductivity: <br>
power factor = $\sigma S^{2}$<br>
It is a measure of the efficiency of a thermoelectric material in converting heat energy into electrical energy. 
The power factor can be plotted with respect to the chemical potential or to the charge carrier concentration with the two functions repoted below. The functions will ask as an input the direction in which you want the electron conductivity to be plotted. Pay attention to the fact that these functions require two input file, the Seebeck and the $\sigma$.

```python
plot_cry_powerfactor_potential(seebeck,sigma)
```

```python
plot_cry_powerfactor_carrier(seebeck,sigma)
```
